# Pittsburgh Air Quality
### This data set provided various names of places in Pittsburgh along with their measured AQI (Air Quality Index) at different times.  Using this data and the code below, I found the average AQI for each area.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
file = 'airQuality.csv'
air_quality_data = pd.read_csv(file)

# Grouping the data by 'site' and calculating the average 'index_value'
average_index_per_site = air_quality_data.groupby('site')['index_value'].mean().reset_index()

# Renaming the 'index_value' column to 'average_index_value'
average_index_per_site.rename(columns={'index_value': 'average_index_value'}, inplace=True)

# Plotting the data as a bar graph
plt.figure(figsize=(10, 6))
plt.bar(average_index_per_site['site'], average_index_per_site['average_index_value'], color='blue')
plt.xlabel('Site')
plt.ylabel('Average Index Value')
plt.title('Average Index Value by Site')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()  
plt.show()
```

### A **lower AQI** indicates cleaner air. Based on the graph, it appears that Lawrenceville is the lowest, but because this data set separates Lawrenceville into two parts, it is actually closer to the median.  This means that the areas with the cleanest air on average are **West Mifflin**, **Flag Plaza**, and **Glassport High Street**.
